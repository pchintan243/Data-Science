```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,39,48,12,23,67]
```

# g means color green and + denotes marker, -- denotes dashed-line

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Dataset")
plt.plot(x,y, '--g+')
```

# alpha is use for transperency

```python
plt.plot(x,y, color='red', marker='D', markersize=12, alpha=0.1)
```
