```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 149, 150, 88, 93, 115, 135, 80, 77, 82, 129]

# bins is use for range
 
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.9, color="g")
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total number of patients")
plt.title("Blood sugar analysis")

blood_sugar_men = [113, 85, 90, 149, 150, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [88, 93, 115, 113, 85, 90, 149, 150, 135, 80, 77, 82, 129]

# bins is use for range
# orientation is use for get the chart in horizontal bar
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.9, color=["green", "red"], label= ["men","women"]
        , orientation='horizontal')

plt.legend()
```
