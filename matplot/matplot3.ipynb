```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max = [50,51,39,48,62,53,67]
min = [32,22,28,34,33,21,34]
avg = [28,32,44,23,12,34,32]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max, color='green', label="max")
plt.plot(days,min, color='red', label="min")
plt.plot(days,avg, color='pink', label="avg")

# legend is use for find the dataset using lable
plt.legend(shadow=True)

# To get the grid
plt.grid()
```
