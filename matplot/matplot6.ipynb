```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home rent", "Food", "Phone", "Car", "Other"]
```

```python
#  equal is use to make the circle as a round
plt.axis("equal")

# radius is increase the size of circle
#  autopct is use for getting the percentage
# explode is use for get the data from outside
# startangle is use for start the data from 45 degree
plt.pie(exp_vals, labels=exp_labels, radius=1.3, autopct="%0.2f%%", explode=[0,0.3,0,0,0], startangle=45)

# show is use for only show the chart
plt.show()
```
