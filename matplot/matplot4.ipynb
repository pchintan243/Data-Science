```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = ['GOGGLE','AMAZON','MS','APPLE']
rev = [90, 124, 52, 454]
profit = [12,45,23,122]
```

```python
xpos = np.arange(len(data))
xpos
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue(us)")
plt.title("Us Tech stock")


# To get the name in horizontal bar
plt.xticks(xpos, data)

# Using this easily find profit and revenue
# +0.2 & -0.2 is perform on our array and give width to show profit and revenue side by side
plt.bar(xpos-0.2, rev, width=0.4, label="rev")
plt.bar(xpos+0.2, profit, width=0.4, label="profit")

plt.legend()
```

```python
# To get the name in vertical bar
plt.yticks(xpos, data)

# barh is use for get the chart on y-axis
plt.barh(xpos-0.2, rev, height=0.4, label="rev")
plt.barh(xpos+0.2, profit, height=0.4, label="profit")

plt.legend()
```
